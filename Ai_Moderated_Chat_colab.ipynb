```python
API_URL = "https://openrouter.ai/api/v1/chat/completions"

# Load environment variables from .env file
load_dotenv()
API_KEY = os.getenv("OPENROUTER_API_KEY")
```

## Moderation Logic

```python
BANNED_KEYWORDS = ["kill", "hack", "bomb"]

def is_safe(text):
    text = text.lower()
    for word in BANNED_KEYWORDS:
        if word in text:
            return False
    return True
```

## system Prompt + API Call

```python
def call_ai(user_prompt):
    system_prompt = (
        "You are a helpful assistant. "
        "You must not provide harmful, illegal, or unethical advice."
    )

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "gpt-3.5-turbo",
        "messages": [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    }

    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()["choices"][0]["message"]["content"]
```

```python
user_prompt = input("Enter your prompt: ")

# Input moderation
if not is_safe(user_prompt):
    print("Your input/output violated the moderation policy.")
else:
    # Check if API_KEY is set before calling AI
    if not API_KEY:
        print("Error: API Key is not set. Please ensure your .env file is configured correctly and OPENROUTER_API_KEY is set.")
    else:
        ai_response = call_ai(user_prompt)

        # Output moderation
        if not is_safe(ai_response):
            print("Your input/output violated the moderation policy.")
        else:
            print("AI Response:")
            print(ai_response)
```
